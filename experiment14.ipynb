```python
import pandas as pd
import numpy as np

data = {
    'Product':['Laptop', 'Mobile', 'Tablet', 'Headphones', 'Smartwatch', 'Camera'],
    'Price':[55000, 20000, 30000, 2000, 8000, 45000],
    'Units_Sold':[25, 60, 35, 80, 50, 15]
}
df= pd.DataFrame(data)
print(df)
```

```python
bins1 = [0, 10000, 30000, 60000]
labels1 = ['Low', 'Medium', 'High']

df['Price_Category'] = pd.cut(df['Price'], bins=bins1, labels=labels1)

print(df)
```

```python
bins2 = [0,20,50,80]
labels2 = ['Low sales', 'Medium sales', 'High sales']
df['Sales_Category'] = pd.cut(df['Units_Sold'], bins=bins2, labels=labels2)
print(df)
```

```python
print(df.dtypes)
```

```python
df['Units_Sold'] = df['Units_Sold'].astype(float)
print(df.dtypes)
```

```python
print(df)
```

```python
df['Product'] = df['Product'].str.upper()
print(df)
```

```python
df['Price'] = df['Price'].round(2)
print(df)
```

```python
sorted_df= df.sort_values(by='Units_Sold', ascending= False)
print(sorted_df)
```

```python
df.sort_values(by='Units_Sold', inplace= True)
print(df)
```

```python
print(df['Price_Category'].unique())
```

```python
#Dataset: Food Delivery Orders

import pandas as pd

import numpy as np

data = {

'Order_ID': ['01', '02', '03', '04', '05', '06', '07', '08'],

'Order_Value': [250,450, 120, 600, 300, 150,500,220],

'Delivery_Time': [30,45,25,60,35,20,50,28],

'Distance_km': [5,8,3,10,6,2,9,4]

}

df= pd.DataFrame(data)

print(df)
```

```python
df['Order_Value'] = df['Order_Value'].astype(float)
print(df.dtypes)
```

```python
df.sort_values(by='Order_Value', inplace= True, ascending= False)
print(df)
```

```python
bin1 = [0,200,400,600]
labels1=['Low', 'Medium', 'High']
df['Order_Category'] = pd.cut(df['Order_Value'], bins=bin1, labels=labels1)
print(df)
```

```python
bin2 = [0,2,6,10]
labels2=['Close', 'Normal', 'Far']
df['Distance_Category'] = pd.cut(df['Distance_km'], bins=bin2, labels=labels2)
print(df)
```

```python
bin2 = [0,20,45,60]
labels2=['Fast', 'Normal', 'Slow']
df['Delivery_Category'] = pd.cut(df['Delivery_Time'], bins=bin2, labels=labels2)
print(df)
```
